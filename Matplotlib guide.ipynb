__Step 1:__
<br> Importing necessary packages and apply formatting

* Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

__Step 2:__
<br>Getting data:

- Creating own data

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
```

<nbsp><nbsp><nbsp><nbsp>Notes:<br>
    * It is possible to use Pandas Dataframe to get data to trend<br>
    * It is possible to use numpy library to create data automatically


__Step 3__
<br>
Making a plot


```python
plt.plot(x, y)
```

__Step 4__
<br> Making trend bigger

```python
plt.figure(figsize = (10, 5)) #new line
plt.plot(x, y)
```

__Step 5.__
<br>Adding title

```python
plt.figure(figsize = (10, 5))
plt.title('Beautiful Trend', size=18) # new line
plt.plot(x, y)
```

__Step 6__
<br>Adding axis titles

```python
plt.figure(figsize = (10, 5))
plt.title('Beautiful Trend', size=18)
plt.xlabel('Horizontal axis') #new line
plt.ylabel('Vertical axis') #new line
plt.plot(x, y)
```

__Step 7__<br>
Adjusting trend scales

```python
plt.figure(figsize = (10, 5))
plt.title('Beautiful Trend', size=18)
plt.xlabel('Horizontal axis')
plt.ylabel('Vertical axis')
plt.axis (xmin = 0, xmax =12, ymin = 0, ymax = 12) #new line
plt.plot(x, y)
```

__Step 8__
<br> Adding more trend types and trend formatting

```python
plt.figure(figsize = (10, 5))
plt.title('Beautiful Trend', size=18)
plt.xlabel('Horizontal axis')
plt.ylabel('Vertical axis')
plt.axis (xmin = 0, xmax =12, ymin = 0, ymax = 12)
plt.plot(x, y, linewidth = 2) #modified line making line wider
plt.bar(y, x, color = 'yellow', alpha = 0.7 ) #new line
plt.scatter(x, y, linewidth = 4, color = 'red', marker = 's') #new line
```

__Step 9__
<br> Adding trend labels and legend

```python
plt.figure(figsize = (10, 5))
plt.title('Beautiful Trend', size=18)
plt.xlabel('Horizontal axis')
plt.ylabel('Vertical axis')
plt.axis (xmin = 0, xmax =12, ymin = 0, ymax = 12)
plt.plot(x, y, linewidth = 2, label = 'line plot') #modified line, added label
plt.bar(y, x, color = 'yellow', alpha = 0.7, label = 'bar chart' ) #modified line, added label
plt.scatter(x, y, linewidth = 4, color = 'red', marker = 's', label = 'scatter plot') #modified line, added label
plt.legend() #new line
```

__Step 10__
<br>Adding grid

```python
plt.figure(figsize = (10, 5))
plt.title('Beautiful Trend', size=18)
plt.xlabel('Horizontal axis')
plt.ylabel('Vertical axis')
plt.axis (xmin = 0, xmax = 12, ymin = 0, ymax = 12)
plt.plot(x, y, linewidth = 2, label = 'line plot')
plt.bar(y, x, color = 'yellow', alpha = 0.7, label = 'bar chart' )
plt.scatter(x, y, linewidth = 4, color = 'red', marker = 's', label = 'scatter plot')
plt.legend(loc ='upper right')
plt.grid (axis = 'both') #new line
```

__PUTTING IT ALL TOGETHER__
<br> Feel free to modify any of the paramters in the cell below to see what happens, add own data, adjust formatting, add more trends, etc.

```python
#Importing necessary packages and apply formatting
import matplotlib.pyplot as plt
%matplotlib inline

#Getting data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

#Changing trend size
plt.figure(figsize = (10, 5))

#Adding title
plt.title('Beautiful Trend', size=18)

#Adding axis titles
plt.xlabel('Horizontal axis')
plt.ylabel('Vertical axis')

#Changing trend scales
plt.axis (xmin = 0, xmax = 12, ymin = 0, ymax = 12)

#Creating line plot
plt.plot(x, y, linewidth = 2, label = 'line plot')

#Creating bar plot
plt.bar(y, x, color = 'yellow', alpha = 0.7, label = 'bar chart' )

#Creating scatter plot
plt.scatter(x, y, linewidth = 4, color = 'red', marker = 's', label = 'scatter plot')

#Adding legend
plt.legend(loc ='upper right')

#Adding grid
plt.grid (axis = 'both')
```
